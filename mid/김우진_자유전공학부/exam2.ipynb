```python
# 출력양식

print('--------------------------------------')
print('')
print('명단 관리 프로그램')
print('')
print('입력양식 : 이름, 키, 몸무게, 좋아하는 브랜드')
print('(중단 : n)')
print('')
print('--------------------------------------')


# 초기 변수 선언

people = {}
i = 1

height_lis = []
weight_lis = []
brand_lis = []


# 제어 구조

while True:
    line = input('>')
    if line.lower() == 'n':
        if i < 10:
            print('10명 이상의 데이터가 입력되어야 합니다.')
            continue
        else:
            break


    ## list add 

    lis = []
    
    try: 
        name, height, weight, brand = line.split(',')
    
        lis.append(name)
        lis.append(int(height))
        lis.append(int(weight))
        lis.append(brand)

        people[i]= lis

        height_lis.append(int(height))
        weight_lis.append(int(weight))
        brand_lis.append(brand)

        i += 1

    except: #양식에 맞지 않을 경우, 다시 입력하게 만든다.
        print ("입력 양식에 맞게 입력하세요.")
        continue
    
    


#계산 함수 설정

def mean(numbers):    #평균값 함수 
    return float(sum(numbers)) / max(len(numbers), 1)

def median(array):    #중위값 함수
    array = sorted(array)
    half, odd = divmod(len(array), 2)
    if odd:
        return array[half]
    return (array[half - 1] + array[half]) / 2.0


# brand list

word_cnt = {}
for word in brand_lis:
    if word in word_cnt:
        word_cnt[word] += 1
    else:
        word_cnt[word] = 1

common_words = sorted(word_cnt, key= word_cnt.get, reverse=True)


#키와 몸무게의 최소값, 최대값, 평균값, 중위값을 출력할 것

print ('')
print ('')
print('--------------------------------------')
print ('')
print ('키, 몸무계의 통계') 
print ('')
print ('키')
print (('최소값: {}'.format(min(height_lis)), '최대값: {}'.format(max(height_lis)), '평균값: {}'.format(mean(height_lis)), '중위값: {}'.format(median(height_lis))))
print ('')
print ('몸무게')
print (('최소값: {}'.format(min(weight_lis)), '최대값: {}'.format(max(weight_lis)), '평균값: {}'.format(mean(weight_lis)), '중위값: {}'.format(median(weight_lis))))


#좋아하는 브랜드 랭킹 TOP3 출력할 것
print ('')
print ('')
print('--------------------------------------')
print ('')
print ('좋아하는 브랜드 랭킹 TOP 3, 좌측부터 1위')
print ('')
print (common_words[:3])
print ('')
print('--------------------------------------')
print ('')
```
