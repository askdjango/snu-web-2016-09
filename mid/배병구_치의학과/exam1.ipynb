```python
def print_diamond(size):

    print('Case #1')
    width = size
    formatter = '{:<' + str(width) + '}'

    for i in range(1, size+1):
        amount = i
        print(formatter.format('*' * amount))


    print('Case #2')
    width = size
    formatter = '{:>' + str(width) + '}'

    for i in range(1, size+1):
        amount = i
        print(formatter.format('*' * amount))


    print('Case #3')
    width = 2 * size - 1
    formatter = '{:^' + str(width) + '}'

    for i in range(1, size+1):
        amount = 2 * i - 1
        print(formatter.format('*' * amount))


    print('Case #4')
    width = 2 * size - 1
    formatter = '{:^' + str(width) + '}'
    for i in range(size, 0, -1):
        amount = 2 * i - 1
        print(formatter.format('*' * amount))


    print('Case #5')
    width = 2 * size - 1
    formatter = '{:^' + str(width) + '}'

    for i in range(1, size+1):
        amount = 2 * i - 1
        print(formatter.format('*' * amount))

    for i in range(size-1, 0, -1):
        amount = 2 * i - 1
        print(formatter.format('*' * amount))


    print('Case #6')
    blank_count = size - 1
    sharp_count = 1
    print(' ' * blank_count, end='')
    print('#' * sharp_count)

    for i in range(2, size+1):
        blank_count = size - i
        sharp_count = 1
        star_count = i * 2 -3
        print(' ' * blank_count, end='')
        print('#' * sharp_count, end='')
        print('*' * star_count, end='')
        print('#' * sharp_count)

    for i in range(size-1, 1, -1):
        blank_count = size - i
        sharp_count = 1
        star_count = i*2 -3
        print(' ' * blank_count, end='')
        print('#' * sharp_count, end='')
        print('*' * star_count, end='')
        print('#' * sharp_count)

    blank_count = size - 1
    sharp_count = 1
    print(' ' * blank_count, end='')
    print('#' * sharp_count)

    print('아래의 None은 무시하세요 :)')

print(print_diamond(size = int(input('> '))))
```
