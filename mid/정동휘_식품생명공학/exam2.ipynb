```python
from collections import defaultdict
import operator

print('입력양식 : 이름, 키, 몸무게, 좋아하는브랜드')
print('(중단 : n),(10명 이상 입력해야 합니다)')
print('')

height_list = []
weight_list = []
brand_count = defaultdict(int)
number = 0

while True:
    line = input('> ')
    if line.lower() == 'n' and number >= 10:
        break
    elif line.lower() == 'n' and (number >= 1 and number < 10):
        print('10명 이상의 데이터가 필요합니다')
    name, height, weight, brand = line.split(',')
    height_list.append(int(height))
    weight_list.append(int(weight))
    brand_count[brand] += 1
    number += 1

    
# 키

height_list.sort()
print('키 최소값: '+str(height_list[0]))
print('키 최대값: '+str(height_list[len(height_list)-1]))
height_sum = 0
for i in range(0,len(height_list)):
    height_sum += height_list[i]
height_mean = height_sum / len(height_list)
print('키 평균값: '+str(height_mean))
if len(height_list) % 2 == 0:
    print('키 중앙값: '+str(height_list[int(len(height_list)/2-1)]))
else:
    print('키 중앙값: '+str(height_list[int((len(height_list)+1)/2-1)]))
    

# 몸무게

weight_list.sort()
print('몸무게 최소값: '+str(weight_list[0]))
print('몸무게 최대값: '+str(weight_list[len(weight_list)-1]))
weight_sum = 0
for i in range(0,len(weight_list)):
    weight_sum += weight_list[i]
weight_mean = weight_sum / len(weight_list)
print('몸무게 평균값: '+str(weight_mean))
if len(weight_list) % 2 == 0:
    print('몸무게 중앙값: '+str(weight_list[int(len(weight_list)/2-1)]))
else:
    print('몸무게 중앙값: '+str(weight_list[int((len(weight_list)+1)/2-1)]))

    
# 브랜드

sorted_brand = sorted(brand_count.items(), key=operator.itemgetter(1), reverse=True)
print('브랜드 TOP1은 '+str(sorted_brand[0][0]))
print('브랜드 TOP2는 '+str(sorted_brand[1][0]))
print('브랜드 TOP3는 '+str(sorted_brand[2][0]))
```
