## 공동순위를 고려한 방법

```python

print('입력양식 : 이름, 키, 몸무게, 좋아하는브랜드')
print('(중단 : n)')
print('')

people = {}

# 키에 관한 list와 total sum 지정
lst_height = []
tot_height = 0
# 몸무게에 관한 list와 totla sum 지정
lst_weight = []
tot_weight = 0

# 좋아하는 브랜드에 대한 tuple 지정
lst_brand = []

# 마지막 Dictionary의 key, value 서로 바꾸기
res = {}

while True:
    line = input('> ')
    if line.lower() =='n':
        break
    
    # 사전(Dictionary)를 이용해서 순서대로 이름, 키, 몸무게, 좋아하는브랜드 입력받기
    
    name, height, weight, brand = line.split(', ') 
    
    # TODO: 이하 구현
    
    people[name] = {'height': height, 'weight': weight, 'brand': brand}
    print(people[name])
    
    tot_height += int(people[name]['height'])
    lst_height.append(int(people[name]['height']))
    
    tot_weight += int(people[name]['weight'])
    lst_weight.append(int(people[name]['weight']))
    
    lst_brand.append(people[name]['brand'])

# 키에 관한 리스트를 .sort()를 이용해 오름차순으로 정렬
lst_height.sort()
print('키의 최소값:', lst_height[0])

# 내림차순으로 정렬
lst_height.sort(reverse=True)
print('키의 최대값:', lst_height[0])

# 평균 구하기(미리 for문에서 총합 tot_height을 이용하여)
avg_height = tot_height / len(people)
print('키의 평균값:', avg_height)

def median(lst_height):
    lst_height.sort()
    
    if (len(people) == 0): 
        return None
    center = int(len(people) / 2)
    if (len(people) % 2 == 1): 
        return int(lst_height[center])
    else: 
        return (int(lst_height[center - 1]) + int(lst_height[center])) / 2.0
    
median(lst_height)
print('키의 중위값:', median(lst_height), '\n')

# 몸무게에 관한 리스트를 .sort()를 이용해 오름차순으로 정렬
lst_weight.sort()
print('몸무게의 최소값:', lst_weight[0])

# 내림차순으로 정렬
lst_weight.sort(reverse=True)
print('몸무게의 최대값:', lst_weight[0])

# 평균 구하기(미리 for문에서 총합 tot_height을 이용하여)
avg_weight = tot_weight / len(people)
print('몸무게의 평균값:', avg_weight)

def median(lst_weight):
    lst_weight.sort()
    
    if (len(people) == 0): 
        return None
    center = int(len(people) / 2)
    if (len(people) % 2 == 1): 
        return int(lst_weight[center])
    else: 
        return (int(lst_weight[center - 1]) + int(lst_weight[center])) / 2.0
    
median(lst_weight)
print('몸무게의 중위값:', median(lst_weight), '\n')


result = {}


for word in lst_brand:
    if word in result:
        result[word] += 1
    else:
        result[word] = 1

print('브랜드별 선호도 결과:', result, '\n')

N = 3

ordered_values = list(set(result.values()))  # 집합 자료형을 만들고 리스트화하여 중복된 값을 없애고 오름차순 배열
ordered_values.reverse()  # 내림차순으로 배열

top_items = ordered_values[:N]  # Value를 기준으로 N까지
counter = 1

print('TOP3')

for item in top_items:
    top_names = [key for key, val in result.items() if val == item]
    print('선호도 랭킹 {counter}:'.format(counter=counter) + ','.join(top_names))
    counter +=1
```

### median 구하는 함수 정의

```python
a = [3, 6, 7, 9]

def median(a):
    b = sorted(a)
    b_len = len(b)
    if (b_len == 0): return None
    b_center = int(b_len / 2)
    if (b_len % 2 == 1): return b[b_center]
    else: return (b[b_center - 1] + b[b_center]) / 2.0
    
median(a)
```
