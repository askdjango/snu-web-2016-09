```python
print('다음 양식에 맞춰 명단을 작성해 주세요')
print('입력양식 : 이름, 키, 몸무게, 좋아하는 브랜드')
print('(중단 : n)')
print('')

list_name=[]
list_height=[]
list_weight=[]
list_brand=[]
dic_brand={}

while True :
    line= input('> ')    
    if line.lower() == 'n':
        break
    
    name,height,weight,brand=line.split(',')
    list_name.append(name)
    list_height.append(height)
    list_weight.append(weight)
    list_brand.append(brand)
    
list_height.sort()
list_weight.sort()

def average(list):
    sum=0
    for i in list :
        sum= sum+ int(i)
    average=sum/len(list)
    return average

num=len(list_name)

if num%2==0:
    height_middle= list_height[int(int(num)/2)] or list_height[int(int((num)/2-1))]
    weight_middle= list_weight[int(int(num)/2)] or list_weight[int(int((num)/2-1))]
else:
    height_middle= list_height[int(int(num)/2)]
    weight_middle= list_weight[int(int(num)/2)]
    
height_element='''
<키>
최대값 : {}
최소값 : {}
평균값 : {}
중위값 : {}'''.format(max(list_height),min(list_height),average(list_height),(height_middle ))

weight_element='''
<몸무게>
최대값 : {}
최소값 : {}
평균값 : {}
중위값 : {}'''.format(max(list_weight),min(list_weight),average(list_weight),(weight_middle ))

print(height_element)
print(weight_element)


for i in list(list_brand):
    if i in dic_brand:
        dic_brand[i] += 1
    else:
        dic_brand[i] = 1

import operator
rank_dic_brand= sorted(dic_brand.items(), key=operator.itemgetter(1),reverse=True)
for i in range(0,3):
    print("좋아하는 브랜드 랭킹 TOP3",i+1,"위",rank_dic_brand[i])

    

      

```
