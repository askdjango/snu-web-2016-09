```python
def people_profile():
    print('입력양식 : 이름, 키, 몸무게, 좋아하는 브랜드')
    print('(중단 : n)')
    print('')
    
    people = {}
    
    while True:
        line = input('> ')
        if line.lower() == 'n':
            break
            
        name, height, weight, brand = line.split(',')
        people[name] = {'height': float(height), 'weight': float(weight), 'brand': brand.strip()}
        
    return people

def median(data):
    
    data = sorted(data)
    n = len(data)
        
    if n%2 == 1:
        return data[n//2]
    else:
        i = n//2
        return (data[i - 1] + data[i])/2
    
def people_analysis(profile):
    height_list = []
    weight_list = []
    brand_list = []
    
    for name in profile:
        height_list.append(profile[name]['height'])
        weight_list.append(profile[name]['weight'])
        brand_list.append(profile[name]['brand'])
        
    height_min = min(height_list)
    height_max = max(height_list)
    height_average = sum(height_list) / len(height_list)
    height_median = median(height_list)
        
    print("키 = 최소값: {}, 최대값: {}, 평균값: {}, 중위값: {}".format(
            height_min, height_max, height_average, height_median))
    
    weight_min = min(weight_list)
    weight_max = max(weight_list)
    weight_average = sum(weight_list) / len(weight_list)
    weight_median = median(weight_list)
        
    print("몸무게 = 최소값: {}, 최대값: {}, 평균값: {}, 중위값: {}".format(
            weight_min, weight_max, weight_average, weight_median))
    print('')
    
    brand_like_count = {}

    for brand in set(brand_list):
        if brand_list.count(brand) in brand_like_count:
            brand_like_count[brand_list.count(brand)] += ', ' + brand
        else:
            brand_like_count[brand_list.count(brand)] = brand
        
    brand_like_count = sorted(brand_like_count.items(), reverse=True)

    print("좋아하는 브랜드 TOP 3")
    for i in range(len(brand_like_count)):
        if i > 2:
            break
        else:
            print("{}위 : {} / 좋아하는 사람 수 : {}".format(i+1, brand_like_count[i][1], brand_like_count[i][0]))
        
    
people_profile = people_profile()  # 유저 정보를 입력받는 부분
print('')
people_analysis(people_profile)  # 키와 몸무게의 최소값, 최대값, 평균값, 중위값 / 좋아하는 브랜드 랭킹 TOP 3 출력
```
