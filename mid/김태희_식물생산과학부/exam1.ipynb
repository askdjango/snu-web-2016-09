```python
size = int(input('size : '))
size_of_diamond = size * 2 - 1

print('\n')
print('Case #1')
for i in range (1, size+1):
    star = i
    space = size - i
    print('*' * star + ' ' * space)
    
print('\n')
print('Case #2')
for i in range (1, size+1):
    star = i
    space = size - i
    print(' ' * space + '*' * star)
    
print('\n')
print('Case #3')
for i in range (1, size+1):
    star = i * 2 - 1
    space = size - i
    print(' ' * space + '*' * star)
    
print('\n')
print('Case #4')
for i in range (1, size+1):
    star = 2 * (size - i) + 1
    space = i - 1
    print(' ' * space + '*' * star)
    
print('\n')
print('Case #5')
format_string = '{:^' + str(size_of_diamond) + '}'

for i in range (1, size+1):
    star = i * 2 - 1
    print(format_string.format('*' * star))
for i in range (size-1, 0, -1):
    star = i * 2 - 1
    print(format_string.format('*' * star))
    
print('\n')
print('Case #6')

for i in range (1, size+1):
    space = size - i
    if i == 1:
        print(' ' * space + '#')
    else:
        star = i * 2 - 3
        print(' ' * space + '#' + '*' * star + '#')
        
for i in range (size-1, 0, -1):
    space = size - i
    if i == 1:
        print(' ' * space + '#')
    else:
        star = i * 2 - 3
        print(' ' * space + '#' + '*' * star + '#')
        
```
