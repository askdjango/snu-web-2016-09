```python
size = input('크기를 입력하세요 : ')

def case1(size):
    print('Case #1')
    for i in range(0, int(size), +1):
        print((i+1)*'*')
    return ''

def case2(size):
    print('Case #2')
    for i in range(0, int(size), +1):
        print((int(size)-1-i)*' '+(i+1)*'*')
    return ''

def case3(size):
    print('Case #3')
    for i in range(0, int(size), +1):
        print((int(size)-1-i)*' '+(2*i+1)*'*')
    return ''

def case4(size):
    print('Case #4')
    for i in range(int(size)-1, -1, -1):
        print((int(size)-1-i)*' '+(2*i+1)*'*')
    return ''

def case5(size):
    print('Case #5')
    for i in range(0, int(size), +1):
        print((int(size)-1-i)*' '+(2*i+1)*'*')
    for i in range(int(size)-2, -1, -1):
        print((int(size)-1-i)*' '+(2*i+1)*'*')
    return ''

def case6(size):
    print('Case #6')
    print((int(size)-1)*' '+'#')
    for i in range(1, int(size), +1):
        print((int(size)-1-i)*' '+'#'+(2*i-1)*'*'+'#')
    for i in range(int(size)-2, 0, -1):
        print((int(size)-1-i)*' '+'#'+(2*i-1)*'*'+'#')
    if(int(size) > 1):
        print((int(size)-1)*' '+'#')
    return ''

print(case1(size))
print(case2(size))
print(case3(size))
print(case4(size))
print(case5(size))
print(case6(size))
```
