```python
print('입력양식 : 이름, 키, 몸무게, 좋아하는브랜드')
print('(중단 : n)')
print('')

people = {}
namelist = []
heightlist = []
weightlist = []
brandlist = []

while True:
    line = input('> ')
    if line.lower() == 'n':
        break

    name, height, weight, brand = line.split(',')
    namelist.append(name)
    heightlist.append(height)
    weightlist.append(weight)
    brandlist.append(brand)

# TODO: 이하 구현

def listresult1(listinput):

    listinputlen = len(listinput)
    listinputsum = 0
    listinputmed = 0
    listinputfloat = []

    for i in range(0, listinputlen):
        listinputfloat.append(float(listinput[i]))

    listinputfloat.sort()

    print('최솟값 : ' + str(min(listinputfloat)))
    print('최댓값 : ' + str(max(listinputfloat)))

    for i in listinput:
        listinputsum += float(i)

    print('평균값 : ' + str(listinputsum/listinputlen))

    if listinputlen % 2 == 0:
        varlen = int(listinputlen * 0.5)
        listinputmed = ( float(listinputfloat[varlen - 1]) + float(listinputfloat[varlen]) ) * 0.5
        print('중위값 : ' + str(listinputmed))
    else:
        varlen = int((listinputlen - 1) * 0.5)
        listinputmed = listinputfloat[varlen]
        print('중위값 : ' + str(listinputmed))

def listresult2(listinput):
    namelist = []
    countlist = []
    listinput.sort()

    while True:
        listinput.count(listinput[0])
        countlist.append(listinput.count(listinput[0]))
        namelist.append(listinput[0])
        for i in range(0, listinput.count(listinput[0])):
            listinput.remove(listinput[0])
        if len(listinput) == 0:
            break

    for j in range(0, 3):
        varmax = max(countlist)

        for i in range(0, len(countlist)):
            if countlist[i] == varmax:
                print(str(j+1)+'위 (' + str(countlist[i]) + '회) : ' + str(namelist[i]))
                countlist[i] = 0

print('-------------')
print('키')
listresult1(heightlist)
print('-------------')
print('몸무게')
listresult1(weightlist)
print('-------------')
print('브랜드')
listresult2(brandlist)
print('-------------')
```
