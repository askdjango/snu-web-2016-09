# 중간고사 과제(2)

```python
print('입력양식 : 이름, 키, 몸무게, 좋아하는브랜드')
print('(중단 : n)')
print('')

people = {}

while True:
    line = input('> ')
    if line.lower() == 'n':
        break

    name, height, weight, brand = line.split(',')
    people[name] = {
        '키': height, 
        '몸무게': weight, 
        '브랜드': brand
    }
    print('{}님의 정보가 추가되었습니다.'.format(name))

def median(lst):
    sort_lst = sorted(lst)
    len_lst = len(lst)
    index = (len_lst - 1) // 2

    if (len_lst % 2) > 0:
        return int(sort_lst[index])
    else:
        return (int(sort_lst[index]) + int(sort_lst[index + 1]))/2.0

def mean(lst):
    total_lst = 0
    len_lst = len(lst)
    for i in range(0, len(lst)):
        total_lst += int(lst[i])
    mean_lst = total_lst / len_lst
    return mean_lst

from collections import defaultdict

height_list = []
weight_list = []
brand_list = []

for person in people:
    height_list.append(people[person]['키'])
    weight_list.append(people[person]['몸무게'])
    brand_list.append(people[person]['브랜드'])

def brand_count(brand_list):
    brand_count = defaultdict(int)
    for brand in brand_list:
        brand_count[brand] += 1
    return brand_count

brand_count= brand_count(brand_list)

def brand_result(brand_count):
    brand_count1 = []
    for item in brand_count.items():
        brand_count1 += [item]
    brand_result = sorted(brand_count1, key=lambda brand:brand[1])
    return brand_result

brand_result = brand_result(brand_count)

def brand_reverse(brand_result):
    brand_reverse = defaultdict(list)
    for items in brand_result:
        brand_reverse[items[1]] += [items[0]]
    return brand_reverse

brand_reverse = brand_reverse(brand_result)

def brand_ranked(brand_reverse):
    keys = []
    for key in brand_reverse.keys():
        keys += [key]
    sorted(keys)
    top1 = brand_reverse[keys[-1]]
    top2 = brand_reverse[keys[-2]]
    top3 = brand_reverse[keys[-3]]
    return 'TOP 3 브랜드\n1. {}\n2. {}\n3. {}\n'.format(top1, top2, top3)

brand_ranked = brand_ranked(brand_reverse)

print('키 최솟값: {}'.format(min(height_list)))
print('키 최댓값: {}'.format(max(height_list)))
print('키 평균값: {}'.format(mean(height_list)))
print('키 중간값: {}'.format(median(height_list)))

print('몸무게 최솟값: {}'.format(min(weight_list)))
print('몸무게 최댓값: {}'.format(max(weight_list)))
print('몸무게 평균값: {}'.format(mean(weight_list)))
print('몸무게 중간값: {}'.format(median(weight_list)))

print(brand_ranked)
```
