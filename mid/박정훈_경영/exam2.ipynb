```python
print('입력양식 : 이름, 키, 몸무게, 좋아하는브랜드')
print('(중단 : n)')
print('')

people = {}
while True:
    line = input('> ')
    if line.lower() == 'n':
        break
        
    name, height, weight, brand = line.split(',')
# 과제 시작
    people[name]={'height':height,'weight':weight,'brand':brand}

# 중앙값 함수 선언
def getMedian(x) :
    x_len = len(x)
    if (x_len == 0) : 
        return none
    x_center = int(x_len / 2)
    
    if (x_len % 2 == 1) :
        return int(x[x_center])
    else :
        return int((x[x_center - 1] + x[x_center])) / 2

# 키의 평균값,최대값,최소값,중앙값 구하기
height_group = []
total_height=0

for name in people :
    height_group.append(int(people[name].get('height')))
    total_height += int(people[name]['height'])

average_height = total_height / len(people)
height_group.sort()

print('키의 평균값 : {}'.format(average_height))
print('키의 최대값 : {}'.format(max(height_group)))
print('키의 최소값 : {}'.format(min(height_group)))
print('키의 중앙값 : {}'.format(getMedian(height_group)))


# 몸무게의 평균값,최대값,최소값,중앙값 구하기
weight_group = []
total_weight=0

for name in people :
    weight_group.append(int(people[name].get('weight')))
    total_weight += int(people[name]['weight'])

average_weight = total_weight / len(people)
weight_group.sort()

print('몸무게의 평균값 : {}'.format(average_weight))
print('몸무게의 최대값 : {}'.format(max(weight_group)))
print('몸무게의 최소값 : {}'.format(min(weight_group)))
print('몸무게의 중앙값 : {}'.format(getMedian(weight_group)))

# 브랜드 랭킹 탑3 구하기
brand_group = []

for name in people :
    brand_group.append(people[name].get('brand'))

from collections import defaultdict

brand_count = defaultdict(int)
for brand in brand_group :
    brand_count[brand] += 1

brand_count_num = list(brand_count.values())
brand_count_num = list(set(brand_count_num))
brand_count_num.reverse()

first = []
second = []
third = []

for key,values in brand_count.items():
    if values == brand_count_num[0]:
        first.append(key)
    if len(brand_count_num) > 1 and values == brand_count_num[1]:
        second.append(key)
    if len(brand_count_num) > 2 and values == brand_count_num[2]:
        third.append(key)

print('1등 브랜드',first)
print('2등 브랜드',second)
print('3등 브랜드',third)
```
