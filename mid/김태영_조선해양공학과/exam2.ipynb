```python
## 중간고사 과제2
## 2011-11894 김태영



print('입력양식 : 이름, 키, 몸무게, 좋아하는브랜드')
print('(중단 : n)')
print('')

people = {}

while True:
    line = input('> ')
    if line.lower() == 'n':
        break

    name, height, weight, brand = line.split(',')
    people[name]={'height': height,'weight': weight,'brand': brand}

    

## 딕셔너리의 data에 대해서 최소/최대/평균/중위값을 구하는 함수 정의
def calculation(data, **ppl):
    
    number = len(ppl)
    list = []
    total = int()
    for name in ppl:
        list.append(ppl[name][data])
        total += int(ppl[name][data])
    list.sort()

    print('최소값?', list[0])
    print('최대값?', list[number-1])
    print('평균값?', total/number)

    if number % 2 == 1:
        print('중위값?', list[int(number/2)])
    else:
        middle=0
        middle= int(list[int(number/2)])+int(list[int(number/2-1)])
        print('중위값?', middle/2)
    return None



##함수를 이용하여 사람들 키의 최소/최대/평균/중위값 출력
print('###사람들의 키###')
calculation('height',**people)
print('')



##함수를 이용하여 사람들 몸무게의 최소/최대/평균/중위값 출력
print('###사람들의 몸무게###')
calculation('weight',**people)
print('')
    
    

##사람들이 좋아하는 브랜드 TOP3 출력
brand_list = []
for name in people:
    brand_list.append(people[name]['brand'])
   
from collections import defaultdict

brand_count = defaultdict(int)
for brand in brand_list:
    brand_count[brand] += 1
    
    
brand_sort = (sorted(zip(brand_count.values(),brand_count.keys())))
top = int(1)
size =len(brand_sort)-1

print('###사람들이 좋아하는 브랜드 TOP3 (브랜드/인원)###')
while top <= 3 and size >= 0:
    if brand_sort[size][0]==brand_sort[size-1][0]:
        while brand_sort[size][0]==brand_sort[size-1][0]:
            print(brand_sort[size][1], '(',brand_sort[size][0],'명 )')
            top += 1
            size -= 1
        print(brand_sort[size][1], '(',brand_sort[size][0],'명 )')
        top += 1
        size -= 1
    else:
        print(brand_sort[size][1], '(',brand_sort[size][0],'명 )')
        top += 1
        size -= 1
```

```python
######### 코딩 해설 



print('입력양식 : 이름, 키, 몸무게, 좋아하는브랜드')
print('(중단 : n)')
print('')

people = {}                                     #people이라는 빈 사전 정의

while True:                                     #n을 눌러서 break를 하기 전까지 사람들의 데이터 수집하여 people이라는 사전에 저장
    line = input('> ')
    if line.lower() == 'n':
        break

    name, height, weight, brand = line.split(',')
    people[name]={'height': height,'weight': weight,'brand': brand}

    

## 딕셔너리의 data에 대해서 최소/최대/평균/중위값을 구하는 함수 정의
def calculation(data, **ppl):                   #calculation이라는 함수 정의. data와 **ppl이라는 두 인자를 받음
    
    number = len(ppl)                           
    list = []  
    total = int()                               #받은 사람 수를 number로 하고 빈 리스트 list와 값이 0인 total 정의
    for name in ppl:
        list.append(ppl[name][data])            #for문을 통해서 list에 ppl의 이름마다 data라는 값 입력,
        total += int(ppl[name][data])           #total에는 data값의 총합을 구한다.        
                                                
    list.sort()                                 #list를 오름차순으로 정렬한다.

    print('최소값?', list[0])                    #list에서 제일 앞에 있는 값이므로 최소값
    print('최대값?', list[number-1])             #list에서 제일 뒤에 있는 값이므로 최대값
    print('평균값?', total/number)               #data들의 총합인 total을 사람 수로 나눠서 평균값 계산

    if number % 2 == 1:                         #중위값은 data의 갯수가 홀수와 짝수일 때 구하는 방법이 다르다.
        print('중위값?', list[int(number/2)])    #홀수일 때는, 가장 가운데 있는 값. int(number/2)는 소수점이하 내림이므로 이값을 계산하면 가운데 값 계산 가능.
    else:                                       #짝수일 때는, 가장 가운데 있는 두 값의 평균.
        middle=0
        middle= int(list[int(number/2)])+int(list[int(number/2-1)])
        print('중위값?', middle/2)               #middle이라는 빈 값을 정의하고 가운데 두값의 합을 거기에 대입 후 1/2한 값 출력
    return None



##함수를 이용하여 사람들 키의 최소/최대/평균/중위값 출력
print('###사람들의 키###')
calculation('height',**people)                  #앞서 구한 calculation함수를 이용해 받은 데이터들에서 키에 대한 값들 출력.
print('')



##함수를 이용하여 사람들 몸무게의 최소/최대/평균/중위값 출력
print('###사람들의 몸무게###')
calculation('weight',**people)                  #위와 마찬가지로 weight 입력하여 몸무게에 대한 값들 출력.
print('')
    
    

##사람들이 좋아하는 브랜드 TOP3 출력
brand_list = []
for name in people:
    brand_list.append(people[name]['brand'])   #brand_list라는 빈 리스트에 사람들이 좋아하는 brand들 입력
   
from collections import defaultdict

brand_count = defaultdict(int)
for brand in brand_list:
    brand_count[brand] += 1                    #brand_list리스트를 이용하여 brand_count 사전에 각각의 brand에 따른 숫자를 대입.
                                               #count하는 방법은 단어 카운트 하는 방법과 동일
    
brand_sort = (sorted(zip(brand_count.values(),brand_count.keys())))   #brand_sort에 zip함수 이용하여,(number,'brand')꼴의 튜플들의 리스트 집합 생성.
top = int(1)                                                          #이 때 value에 따른 오름차순으로 정리.
size =len(brand_sort)-1                        #순위 몇번째인지를 나타내는 top, brand_sort의 길이 나타내는 size 정의.

print('###사람들이 좋아하는 브랜드 TOP3 (브랜드/인원)###')
while top <= 3 and size >= 0:                  #top3니까 3번째 이하, size가 0보다 작아지면 중복이 생기므로 제외.
    if brand_sort[size][0]==brand_sort[size-1][0]:               #순위가 공동일 때: 가장 오른쪽(value가 가장 큰 값)에 있는 value값과 두번째의 value 값이 값이 같을때,
        while brand_sort[size][0]==brand_sort[size-1][0]:        #위의 조건을 만족할 때 while문 돌린다.
            print(brand_sort[size][1], '(',brand_sort[size][0],'명 )')                           #이유는 value값이 같으면 공동 순위를 표현해 주어야하기 때문.
            top += 1
            size -= 1                                            #가장 오르쪽에 있는 key값 출력 후에 순위+1,위치 왼쪽1칸 이동.
        print(brand_sort[size][1], '(',brand_sort[size][0],'명 )')
        top += 1
        size -= 1                                                #while문을 빠져나왔을때 value값이 같은 key들 중에 가장 왼쪽에 있는 key는 출력이 안되었기에 이를 출력.
    else:                                     #순위가 단독일 때:
        print(brand_sort[size][1], '(',brand_sort[size][0],'명 )')
        top += 1
        size -= 1                             #brand(key값) 출력 후 마찬가지로 순위+1,위치 왼쪽1칸 이동.
     
```
